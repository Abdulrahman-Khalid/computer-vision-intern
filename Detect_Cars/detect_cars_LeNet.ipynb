```python
import numpy as np
import torch
from torch import nn, optim
import torch.nn.functional as F
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
import math
from typing import List
import argparse
import glob
import shutil
import time
%matplotlib inline
```

```python
def train_and_validate(model, loss_criterion, optimizer, data_set, epochs=25):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    start = time.time()
    model = model.to(device)
    for epoch in range(epochs):
        epoch_start = time.time()
        print(f'Epoch : {epoch+1}/{epochs}')
        model.train()
        train_loss = 0.0 
        train_acc = 0.0
        for i, (inputs, labels) in enumerate(data_set['train']):
            inputs = inputs.to(device)
            labels = labels.to(device)
            # Clean existing gradients
            optimizer.zero_grad()
            # Forward pass - compute outputs on input data using the model
            outputs = model(inputs)
            # Compute loss
            loss = loss_criterion(outputs, labels)
            # Backpropagate the gradients
            loss.backward()
            # Update the parameters
            optimizer.step()
            # Compute the total loss for the batch and it to train_loss
            train_loss += loss.item() * inputs.size(0)
            # Compute the accuracy
            ret, predictions = torch.max(outputs.data, 1)
            correct_counts = predictions.eq(labels.data.view_as(predictions))

            # Convert correct_counts to float and then compute the mean
            acc = torch.mean(correct_counts.type(torch.FloatTensor))
            # Compute total accuracy in the whole batch and add to train_acc
            train_acc += acc.item() * inputs.size(0)
            print(f'Batch number: {i}, Training: Loss: {loss.item()}, Accuracy: {acc.item()}')
```

```python
def count_frames_manual(video):
    total = 0
    while(video.isOpened()):
        ret, _ = video.read()
        if ret == False:
            break
        total +=1
    return total
    
def count_frames(path, override=False):
    video = cv2.VideoCapture(path)
    total = 0
    if override:
        total = count_frames_manual(video)
    else:
        try:
            total = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        except:
            total = count_frames_manual(video)
    video.release()
    return total
```

```python
class LeNet(nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, (5,5), padding=2)
        self.conv2 = nn.Conv2d(6, 16, (5,5))
        self.fc1   = nn.Linear(31616, 120)
        self.fc2   = nn.Linear(120, 84)
        self.fc3   = nn.Linear(84, 2)
    def forward(self, x):
        x = F.max_pool2d(F.relu(self.conv1(x)), (2,2))
        x = F.max_pool2d(F.relu(self.conv2(x)), (2,2))
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
def split_files(video_path, oldpath, cars_files_list, no_cars_files_list, newpath):
    total_number_of_frames = count_frames(video_path)
    l = len(str(total_number_of_frames))
    cars_files = []
    for i in cars_files_list:
        l2 = len(str(i))
        cars_files.append('street'+(l-l2)*'0'+str(i)+'.jpg')
    no_cars_files = []
    for i in no_cars_files_list:
        l2 = len(str(i))
        no_cars_files.append('street'+(l-l2)*'0'+str(i)+'.jpg')
    full_dir = os.path.join(os.getcwd(), f"{oldpath}")
    copy_files(cars_files, full_dir, f"{newpath}/train/cars")
    copy_files(no_cars_files, full_dir, f"{newpath}/train/no_cars")
    

def copy_files(files, old_dir, new_dir):
    new_dir = os.path.join(os.getcwd(), new_dir);
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
    for file in files:
        old_file_path = os.path.join(os.getcwd(), f"{old_dir}/{file}")
        new_file_path = os.path.join(os.getcwd(), f"{new_dir}/{file}")
        shutil.copy(old_file_path, new_file_path)
```

```python
import torchvision.datasets as datasets
import torch.utils.data as loader
from torchvision import transforms

batch_size = 4

image_transforms = {
    'train': transforms.Compose([
              transforms.RandomRotation(degrees=15),
              transforms.RandomHorizontalFlip(),
              transforms.ToTensor(),
              transforms.Normalize([0.485, 0.456, 0.406],[0.229, 0.224, 0.225])
    ]),
}

data = {
    'train': datasets.ImageFolder(root='Data/train', transform=image_transforms['train']),
}

data_set =  {
    'train': loader.DataLoader(data['train'], batch_size=batch_size, shuffle=True),
}
```

```python
loss_criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
train_and_validate(model, loss_criterion, optimizer, data_set, epochs=15)
```

```python
def load_all_frames(frames_path):
    frames = []
    l = os.listdir(frames_path)
    l = sorted(l)
    for file in l:
        if file.endswith('.jpg'):
            frames.append(frames_path +'/'+ file)
    return frames
```

```python
def test_model(model, use_gpu=True):
    device = torch.device('cuda' if torch.cuda.is_available() and use_gpu else 'cpu')
    start = time.time()
    model.eval()
    model = model.to(device)
    font = cv2.FONT_HERSHEY_PLAIN 
    frames = []
    time_list = []
    # new_paths_frames = []
    for idx, frame_path in enumerate(all_frames):
        frame = cv2.imread(frame_path)
        ###########Start##################
        frame_start_time = time.time_ns()
        img = torch.tensor(frame)
        img = img.unsqueeze(0)
        img = img.permute(0,3,1,2)
        img = img.float()
        img = img.to(device)
        with torch.no_grad():
            outputs = model(img)
            prediction = outputs.argmax(axis=1)
            now = time.time_ns()
            model_time = (now - frame_start_time) * 10e-9
            time_list.append(model_time)
            ############END#################
            text = 'Cars' if prediction.data == 0 else 'No Cars'
            cv2.putText(frame,  
                        text,  
                        (20, 20),  
                        font, 1,  
                        (255, 255, 0),  
                        1,
                        cv2.LINE_4)
            frames.append(frame)
    return frames, time_list
```

```python
frames, time_list = test_model(model)
time_list = np.array(time_list)
print("################# GPU #################")
print('Model time mean =', np.mean(time_list, axis=0))
print('Model time std =', np.std(time_list, axis=0))
```

```python
frames, time_list = test_model(model, use_gpu=False)
print("################# CPU #################")
print('Model time mean =', np.mean(time_list, axis=0))
print('Model time std =', np.std(time_list, axis=0))
```
